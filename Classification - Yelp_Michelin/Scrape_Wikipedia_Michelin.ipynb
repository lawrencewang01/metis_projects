```python
from __future__ import print_function, division
import pandas as pd
import requests
import pickle

from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_Michelin_starred_restaurants_in_New_York_City'
```

```python
response = requests.get(url)
page = response.text

soup = BeautifulSoup(page,"lxml")
```

### Stars in the table weren't in string format, so Beautiful Soup was needed to scrape it

```python
tables = soup.find_all("table")
tables
```

```python
tables[0]
```

```python
rows = [row for row in tables[0].find_all('tr')]
```

```python
rows[1].find_all('td')[1].find('a')['title']
```

```python
rows[1]
```

```python
len(rows)
```

```python


rows_temp = rows[1:] 
restaurants = {}

for row in rows_temp:
    temp_list = []
    items = row.find_all('td')
    try:
        stars = items[0].find('a')['href']
    except:
        stars = 'temp'


    for i in items:
        try:
            temp_list.append(i.find('a')['href'])
        except:
            temp_list.append('')
    
    restaurants[stars] = temp_list
restaurants
```

```python
df = pd.DataFrame.from_dict(restaurants, orient = 'index')
```

```python
df.head(10)
```

```python
df.columns = ['Name', 'Borough', '2006', '2007', '2008', '2009', '2010', 
                               '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
```

```python
df.tail(20)
```

```python
df = df.reset_index()
```

```python
df.drop('index', axis=1, inplace = True)
```

```python
with open('scraped_stars', 'wb') as to_write:
    pickle.dump(df, to_write)
```

```python
df.to_csv('michelin')
```
