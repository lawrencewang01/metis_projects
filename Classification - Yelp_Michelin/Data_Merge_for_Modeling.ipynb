```python
import pandas as pd
import pickle
import numpy as np
```

```python
with open('yelp_data','rb') as read_file:
    df_yelp = pickle.load(read_file)
    
with open('michelin_df','rb') as read_file:
    df_mich = pickle.load(read_file)
```

```python
df_yelp.head()
```

```python
df_yelp = df_yelp.drop('city_counts', axis=1)
```

```python
df_mich.head()
```

```python
df_mich.columns = df_mich.columns.str.strip()
```

```python
df_mich_select = df_mich.loc[:,['Alcohol', 'NoiseLevel', 'OutdoorSeating', 'Attire', 'GoodforGroups',
                         'Range', 'TakesReservations', 'Wi-Fi', 'name', 'review_count', 'stars']]
```

```python
df_yelp.columns = ['alcohol', 'noise', 'outdoor_seating', 'attire', 'good_for_groups',
                         'pricerange', 'takes_reservations', 'wifi', 'city', 'name', 'review_count', 'stars']
```

```python
df_yelp = df_yelp.drop('city', axis = 1)
```

```python
df_mich_select.columns = ['alcohol', 'noise', 'outdoor_seating', 'attire', 'good_for_groups',
                         'pricerange', 'takes_reservations', 'wifi', 'name', 'review_count', 'stars']
```
