```python
data = open('/Users/lawrencewang/Downloads/yelp_dataset/business.json')
```

```python
import ast
from sklearn.feature_extraction.text import CountVectorizer
import json,urllib.request
from pprint import pprint
import pandas as pd
import pickle
import numpy as np
```

```python
counter = 0
mylist = []
errorlines = []
for line in data:
    try:
        line = ast.literal_eval(line)
        mylist.append(line)
    except:
        errorlines.append(line)
    counter += 1
```

```python
df = pd.read_json(json.dumps(mylist))
```

```python
df.head(3)
```

```python
df.columns
```

```python
df['categories'] = df.categories.str.replace(' ', '')
```

```python
df_attributes = pd.io.json.json_normalize(df.iloc[:,1])
df_attributes
```

```python
df_attributes.iloc[0,3]
```

```python
df_attributes.info()
```

#### Now we have to get rid of the features that don't have enough information

```python
# merging my restaurants with their respective attributes

df_merge = pd.merge(df, df_attributes, left_index=True, right_index=True)
```

```python
df_merge.head()
```

```python
# considering only restaurants with 3 stars or more. this will also alleviate imbalanced classes a little

df_merge = df_merge.loc[df_merge.stars >= 3.0]
```

```python
df_merge = df_merge.dropna(subset = ['RestaurantsPriceRange2'])
```

```python
df_merge = df_merge.dropna(subset = ['Ambience'])
```

```python
df_merge.columns = df_merge.columns.str.strip()
```

```python
df_merge = df_merge.loc[:,['Alcohol', 'Ambience', 'BikeParking', 'BusinessAcceptsCreditCards', 'BusinessParking',
                       'Caters', 'GoodForKids', 'GoodForMeal', 'HasTV', 'NoiseLevel', 'OutdoorSeating', 
                       'RestaurantsAttire', 'RestaurantsDelivery', 'RestaurantsGoodForGroups', 'RestaurantsPriceRange2',
                       'RestaurantsReservations', 'RestaurantsTakeOut', 'WiFi', 'city', 'name', 'review_count', 'stars']]
```

```python
df_yelp_final = df_merge.drop(['Caters', 'GoodForMeal', 'GoodForKids', 'HasTV', 'RestaurantsDelivery'], axis =1)
```

```python
df_yelp_final = df_yelp_final.dropna(subset = ['Alcohol', 'Ambience', 'BusinessAcceptsCreditCards', 'NoiseLevel',
                                              'RestaurantsPriceRange2'])
```

```python
df_yelp_final = df_yelp_final.dropna()
df_yelp_final.info()
```

#### Some features were dropped using high level knowledge on whether or not they would impact a Michelin rating, such as HasTV, catering, delivery, etc...

```python
df_yelp_final.drop('BusinessParking', axis=1, inplace=True)
```

```python
city_counts = df_yelp_final['city'].value_counts().rename('city_counts')

df_yelp_final = df_yelp_final.merge(city_counts.to_frame(),
                                left_on='city',
                                right_index=True)
```

```python
df_yelp_final = df_yelp_final.loc[df_yelp_final['city_counts'] > 500]
```

```python
df_yelp_final = df_yelp_final.loc[df_yelp_final.review_count > 50]
```

```python
df_yelp_final.columns
```

```python
df_yelp_final.info()
```

### Cleaning up the values of my columns

```python
df_final_cleaning = df_yelp_final.copy()
```

```python
df_final_cleaning.NoiseLevel = df_final_cleaning.NoiseLevel.apply(lambda x: x.replace("'", ""))
df_final_cleaning.NoiseLevel = df_final_cleaning.NoiseLevel.apply(lambda x: x.replace("uvery_loud", "very_loud"))
df_final_cleaning.NoiseLevel = df_final_cleaning.NoiseLevel.apply(lambda x: x.replace("uaverage", "average"))
df_final_cleaning.NoiseLevel = df_final_cleaning.NoiseLevel.apply(lambda x: x.replace("uquiet", "quiet"))
df_final_cleaning.NoiseLevel = df_final_cleaning.NoiseLevel.apply(lambda x: x.replace("uloud", "loud"))
```

```python
df_final_cleaning.Alcohol = df_final_cleaning.Alcohol.apply(lambda x: x.replace("'", ""))
df_final_cleaning.Alcohol = df_final_cleaning.Alcohol.apply(lambda x: x.replace("ubeer_and_wine", "beer_and_wine"))
df_final_cleaning.Alcohol = df_final_cleaning.Alcohol.apply(lambda x: x.replace("ufull_bar", "full_bar"))
df_final_cleaning.Alcohol = df_final_cleaning.Alcohol.apply(lambda x: x.replace("unone", "none"))
```

```python
df_final_cleaning.WiFi = df_final_cleaning.WiFi.apply(lambda x: x.replace("'", ""))
df_final_cleaning.WiFi = df_final_cleaning.WiFi.apply(lambda x: x.replace("ufree", "free"))
df_final_cleaning.WiFi = df_final_cleaning.WiFi.apply(lambda x: x.replace("uno", "no"))
df_final_cleaning.WiFi = df_final_cleaning.WiFi.apply(lambda x: x.replace("upaid", "paid"))
df_final_cleaning.WiFi = df_final_cleaning.WiFi.apply(lambda x: x.replace("None", "no"))
```

```python
df_final_cleaning.BusinessAcceptsCreditCards.value_counts()
```

```python
df_final_cleaning = df_final_cleaning.drop(df_final_cleaning[df_final_cleaning.BusinessAcceptsCreditCards == 'None'].index)
```

```python
df_final_cleaning = df_final_cleaning.drop(df_final_cleaning[df_final_cleaning.RestaurantsAttire == 'None'].index)
```

```python
df_final_cleaning = df_final_cleaning.drop(df_final_cleaning[df_final_cleaning.RestaurantsReservations == 'None'].index)
```

```python
df_final_cleaning = df_final_cleaning.drop(df_final_cleaning[df_final_cleaning.RestaurantsTakeOut == 'None'].index)
```

```python
df_final_cleaning.info()
```

```python
df_final_cleaning.Alcohol = df_final_cleaning.Alcohol.apply(lambda x: x.replace("None", "none"))
```

```python
df_final_cleaning = df_final_cleaning.drop(df_final_cleaning[df_final_cleaning.NoiseLevel == 'None'].index)
```

```python
df_final_cleaning = df_final_cleaning.drop(df_final_cleaning[df_final_cleaning.OutdoorSeating == 'None'].index)
```

```python
df_final_cleaning.RestaurantsTakeOut.value_counts()
```

```python
df_final_cleaning = df_final_cleaning.drop(df_final_cleaning[df_final_cleaning.WiFi == 'paid'].index)
```

```python
df_final_cleaning.head(3)
```

```python
# final feature selection - dropping unnecessary categoricals

df_final_cleaning = df_final_cleaning.drop(['Ambience', 'BusinessAcceptsCreditCards', 'BikeParking', 'BusinessAcceptsCreditCards', 'RestaurantsTakeOut'], axis = 1)
```

```python
df_final_cleaning.head(3)
```

```python
with open('yelp_data', 'wb') as to_write:
    pickle.dump(df_final_cleaning, to_write)
```
